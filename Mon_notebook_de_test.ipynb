```python
import pickle as pkl
import numpy as np
import bcolz
import os.path
import random
```

```python
from Embeddings import extract_aol_queries
from pyserini.search import pysearch
```

```python
queries=extract_aol_queries('./Data/AOL_queries/')
```

```python
searcher = pysearch.SimpleSearcher('./Data/index-robust04-20191213/')
searcher.set_bm25_similarity(0.9, 0.4)
searcher.set_rm3_reranker(20, 20, 0.5)
```

```python
dataset=[['Qid','doc1id','doc2id','label']]
```

```python
for qid in range(len(queries)):
    ranking=searcher.search(queries[qid],50)
    if(len(ranking)>20):
        samples1=random.choices(list(np.arange(10)),k=10)
        samples2=random.choices(list(np.arange(20,len(ranking))),k=10)
        for i in range(10):
            if(i%2==0):
                dataset.append([qid,samples1[i],samples2[i],1])
            else:
                dataset.append([qid,samples2[i],samples1[i],-1])
        
        
```
